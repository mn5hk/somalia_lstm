```python
import pandas as pd

df = pd.read_csv("../../data/somalia_combined_df.csv")
df.set_index('date', inplace=True)
```

```python
autocorr_lags = dict()
for col in df.columns:
    autocorr_lags[col] = []
    for lag in range(1,366):
        autocorr_lags[col].append(df[col].autocorr(lag))
```

```python
autocorr_lags_df = pd.DataFrame(autocorr_lags)
autocorr_lags_df.plot()
# autocorr_lags.plot()
```
